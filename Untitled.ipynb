```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
pio.renderers.default = 'iframe'
np.random.seed(42)

n = 1200

miasta = {
    "Warszawa":   (52.2297, 21.0122, 1.00),
    "Kraków":     (50.0647, 19.9450, 0.75),
    "Wrocław":    (51.1079, 17.0385, 0.65),
    "Poznań":     (52.4064, 16.9252, 0.55),
    "Gdańsk":     (54.3520, 18.6466, 0.55),
    "Łódź":       (51.7592, 19.4560, 0.50),
    "Katowice":   (50.2649, 19.0238, 0.45),
    "Lublin":     (51.2465, 22.5684, 0.30),
    "Białystok":  (53.1325, 23.1688, 0.25),
    "Szczecin":   (53.4285, 14.5528, 0.35),
}

gatunki = ["rock", "pop", "hip-hop", "electronic", "jazz",
           "classical", "folk", "metal", "indie", "reggae"]

typy_obiektow = ["klub", "arena", "stadion", "festiwal", "teatr", "amfiteatr"]
kapacjety = {
    "klub": (200, 1500), "arena": (3000, 15000), "stadion": (20000, 70000),
    "festiwal": (10000, 80000), "teatr": (400, 2000), "amfiteatr": (1500, 8000),
}
cena_bazowa = {
    "rock": 150, "pop": 200, "hip-hop": 180, "electronic": 160, "jazz": 130,
    "classical": 110, "folk": 90, "metal": 140, "indie": 100, "reggae": 110,
}
cena_mnoznik = {
    "klub": 0.7, "arena": 1.3, "stadion": 1.8,
    "festiwal": 1.5, "teatr": 1.2, "amfiteatr": 1.0,
}

start_date = datetime(2024, 1, 1)
daty = [start_date + timedelta(days=int(d)) for d in np.random.randint(0, 730, n)]

wagi = np.array([miasta[m][2] for m in miasta])
miasto = np.random.choice(list(miasta.keys()), n, p=wagi / wagi.sum())
gatunek = np.random.choice(gatunki, n)
typ_obiektu = np.random.choice(typy_obiektow, n,
                                p=[0.40, 0.15, 0.05, 0.10, 0.15, 0.15])

pojemnosc = np.array([np.random.randint(*kapacjety[t]) for t in typ_obiektu])
wypelnienie = np.clip(np.random.beta(5, 2, n), 0.15, 1.0)
sprzedane = (pojemnosc * wypelnienie).astype(int)

cena = np.array([cena_bazowa[g] * cena_mnoznik[t] for g, t in zip(gatunek, typ_obiektu)])
cena = np.round(cena * np.random.uniform(0.7, 1.4, n), -1)
przychod = (cena * sprzedane).astype(int)

df = pd.DataFrame({
    "event_id": range(50001, 50001 + n),
    "data": daty,
    "miasto": miasto,
    "latitude": [miasta[m][0] for m in miasto],
    "longitude": [miasta[m][1] for m in miasto],
    "gatunek": gatunek,
    "typ_obiektu": typ_obiektu,
    "pojemnosc": pojemnosc,
    "bilety_sprzedane": sprzedane,
    "cena_biletu_pln": cena,
    "przychod_pln": przychod,
})

df.to_csv("koncerty_polska.csv", index=False)
print(f"Wygenerowano plik 'koncerty_polska.csv' — {len(df)} koncertów")



# ==========================================
# CZĘŚĆ 1: Wczytanie i eksploracja
# ==========================================
df = pd.read_csv("koncerty_polska.csv", parse_dates=["data"])

print("--- CZĘŚĆ 1: EKSPLORACJA ---")
print(f"Kształt zbioru: {df.shape}")
print(f"Liczba unikalnych miast: {df['miasto'].nunique()}")
print(f"Liczba unikalnych gatunków: {df['gatunek'].nunique()}\n")

```

```python
# ==========================================
# CZĘŚĆ 1: Wczytanie i eksploracja
# ==========================================
df = pd.read_csv("koncerty_polska.csv", parse_dates=["data"])

print("--- CZĘŚĆ 1: EKSPLORACJA ---")
print(f"Kształt zbioru: {df.shape}")
print(f"Liczba unikalnych miast: {df['miasto'].nunique()}")
print(f"Liczba unikalnych gatunków: {df['gatunek'].nunique()}\n")
```

```python
# ==========================================
# CZĘŚĆ 2: Wykres słupkowy
# ==========================================
df_przychod_miasta = df.groupby("miasto")["przychod_pln"].sum().reset_index()
df_przychod_miasta = df_przychod_miasta.sort_values(by="przychod_pln", ascending=False)

fig_bar = px.bar(
    df_przychod_miasta, x="miasto", y="przychod_pln",
    title="Łączny przychód z koncertów według miast",
    labels={"miasto": "Miasto", "przychod_pln": "Przychód [PLN]"},
    color="przychod_pln", text_auto=".2s"
)
fig_bar.update_layout(showlegend=False, template="plotly_white")
fig_bar.show()
```

```python
# ==========================================
# CZĘŚĆ 3: Wykresy liniowe (szereg czasowy)
# ==========================================
df["miesiac"] = df["data"].dt.to_period("M").astype(str)

# Wykres 1
df_miesiac = df.groupby("miesiac").size().reset_index(name="liczba_koncertow").sort_values("miesiac")
fig_line1 = px.line(df_miesiac, x="miesiac", y="liczba_koncertow", title="Liczba koncertów wg miesięcy", markers=True)
fig_line1.update_layout(template="plotly_white")
fig_line1.show()

# Wykres 2
df_miesiac_typ = df.groupby(["miesiac", "typ_obiektu"]).size().reset_index(name="liczba_koncertow").sort_values("miesiac")
fig_line2 = px.line(df_miesiac_typ, x="miesiac", y="liczba_koncertow", color="typ_obiektu", title="Liczba koncertów wg typu obiektu", markers=True)
fig_line2.update_layout(template="plotly_white")
fig_line2.show()
```

```python

# ==========================================
# CZĘŚĆ 4: Histogram i boxplot
# ==========================================
fig_hist = px.histogram(df, x="cena_biletu_pln", nbins=50, title="Rozkład cen biletów", color_discrete_sequence=["indianred"])
fig_hist.update_layout(template="plotly_white")
fig_hist.show()

fig_box = px.box(df, x="typ_obiektu", y="przychod_pln", color="typ_obiektu", title="Przychody wg typu obiektu")
fig_box.update_layout(template="plotly_white", showlegend=False)
fig_box.show()

print("--- WNIOSEK DO CZĘŚCI 4 ---")
print("Stadiony i festiwale przynoszą największe zyski, bo mieści się tam najwięcej ludzi")
```

```python

# ==========================================
# CZĘŚĆ 5: Scatter plot
# ==========================================
df["wypelnienie"] = df["bilety_sprzedane"] / df["pojemnosc"]

fig_scatter = px.scatter(
    df, x="cena_biletu_pln", y="wypelnienie", color="gatunek", size="pojemnosc",
    hover_name="miasto", hover_data=["typ_obiektu"], title="Wypełnienie sali vs Cena biletu"
)
fig_scatter.update_layout(template="plotly_white")
fig_scatter.show()

print("--- WNIOSEK DO CZĘŚCI 5 ---")
print("Cena biletu nie ma wpływu na wypełnienie sali. Drogie i tanie koncerty wyprzedają się podobnie")
```

```python
# ==========================================
# CZĘŚĆ 6: Mapa
# ==========================================
df_mapa = df.groupby(["miasto", "latitude", "longitude"]).agg(
    srednia_cena=("cena_biletu_pln", "mean"),
    liczba_koncertow=("event_id", "count"),
    laczny_przychod=("przychod_pln", "sum")
).reset_index()
df_mapa["srednia_cena"] = df_mapa["srednia_cena"].round(2)

fig_map = px.scatter_mapbox(
    df_mapa, lat="latitude", lon="longitude", size="liczba_koncertow", color="srednia_cena",
    color_continuous_scale="RdBu_r", hover_name="miasto",
    hover_data=["liczba_koncertow", "srednia_cena", "laczny_przychod"],
    zoom=5, center={"lat": 52.0, "lon": 19.0}, mapbox_style="open-street-map", height=600,
    title="Mapa koncertów w Polsce"
)
fig_map.show()
```

```python

# ==========================================
# CZĘŚĆ 7: Subploty 2x2
# ==========================================
fig_sub = make_subplots(rows=2, cols=2, subplot_titles=("Przychód wg miast", "Liczba wg gatunku", "Rozkład cen", "Wypełnienie wg obiektu"))

fig_sub.add_trace(go.Bar(x=df_przychod_miasta["miasto"], y=df_przychod_miasta["przychod_pln"], marker_color="royalblue"), row=1, col=1)
df_gatunki = df["gatunek"].value_counts().reset_index()
fig_sub.add_trace(go.Bar(x=df_gatunki["gatunek"], y=df_gatunki["count"], marker_color="seagreen"), row=1, col=2)
fig_sub.add_trace(go.Histogram(x=df["cena_biletu_pln"], nbinsx=50, marker_color="indianred"), row=2, col=1)
fig_sub.add_trace(go.Box(x=df["typ_obiektu"], y=df["wypelnienie"], marker_color="goldenrod"), row=2, col=2)

fig_sub.update_layout(height=800, showlegend=False, title_text="Mini-dashboard rynkowy", template="plotly_white")
fig_sub.show()
```
